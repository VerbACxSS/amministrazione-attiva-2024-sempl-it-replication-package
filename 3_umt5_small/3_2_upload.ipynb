# Training

## Dependencies and imports

```python
import os

from transformers import AutoTokenizer
from transformers import AutoModelForSeq2SeqLM
```

```python
HF_TOKEN = os.getenv("HF_TOKEN")
```

## Load trained model

```python
tokenizer = AutoTokenizer.from_pretrained("./epochs/checkpoint-33030")
model = AutoModelForSeq2SeqLM.from_pretrained("./epochs/checkpoint-33030")
```

## Upload

```python
tokenizer.push_to_hub("VerbACxSS/sempl-it-umt5-small", token=HF_TOKEN)
model.push_to_hub("VerbACxSS/sempl-it-umt5-small", token=HF_TOKEN)
```
